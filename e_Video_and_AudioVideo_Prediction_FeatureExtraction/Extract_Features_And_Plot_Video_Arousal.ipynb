```python
import timeit
import os
import sys
sys.path.insert(0, '/media/rb/Omega/0. Thesis Linux/TDD Code/caffe-action_recog/python') #To Look For .py Files in this directory
import caffe
import scipy.io
import scipy as sp
import numpy as np
from pylab import *
from sklearn.metrics import mean_squared_error #For RMSE Calculation
from math import sqrt                          #For RMSE Calculation
import pickle                                  #For saving predicted data externally
import matplotlib.pyplot as plt
from matplotlib import gridspec
%matplotlib inline
print 'Done'
```

```python
testnum = '31';

tic = timeit.default_timer();
model = 'deploy_video_Arousal.prototxt';
thousand = '15';
weights = 'outputModel_video_iter_' + thousand + '000_Arousal.caffemodel';   

net = caffe.Net(model,weights,caffe.TEST);

with open( 'List_' + testnum + '_Video_Arousal.txt', 'r' ) as T :
    path_list = T.readlines()

predicted_value_arr = np.zeros((len(path_list), 1),dtype="f4")
actual_value_arr = np.zeros((len(path_list), 1),dtype="f4")

if not os.path.exists('VideoFeaturesArousal'):
    os.makedirs('VideoFeaturesArousal')
    
feature_list_file = open( 'VideoFeaturesArousal/FeatureList_' + testnum + '_Video_Arousal.txt', 'w') 
for i,l in enumerate(path_list):
    sp = l.split(' ')
    img = caffe.io.load_image(sp[0], color=False )
    img = caffe.io.resize( img, (128, 128, 1) )
    img = np.transpose(img, (2,0,1))
    net.blobs['data_v'].data[...] = img
    
    out = net.forward()
    
    feature_array = net.blobs['fc1_v'].data
    predicted_value = net.blobs['fc2_v'].data
    actual_value = float(sp[1])
    
    predicted_value_arr[i] = predicted_value
    actual_value_arr[i] = actual_value
    
    folddir = sp[0].split('/') 
    directory = 'VideoFeaturesArousal/' + folddir[1];
    filename = directory + '/Video' + filter(str.isdigit, folddir[2]) + 'FeatureArousal.mat';
    if not os.path.exists(directory):
        os.makedirs(directory)
    scipy.io.savemat(filename, mdict={'feature_array':feature_array,'predicted_value':predicted_value,'actual_value':actual_value})
    feature_list_file.write(filename + '\n')
feature_list_file.close()

plt.plot(arange(len(predicted_value_arr)), predicted_value_arr, 'b',arange(len(actual_value_arr)), actual_value_arr, 'r')

rmse = sqrt(mean_squared_error(actual_value_arr, predicted_value_arr))
print rmse

scipy.io.savemat('actual_value_video' + testnum + '_Arousal.mat', mdict={'a_value': actual_value_arr})
scipy.io.savemat('predicted_value_video' + testnum + '_' + thousand + 'k_Arousal.mat', mdict={'p_value': predicted_value_arr})

print 'Complete'
toc = timeit.default_timer();
time_elapsed = toc - tic; #elapsed time in seconds
print time_elapsed
```
